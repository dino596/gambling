  <script>
    document.addEventListener('DOMContentLoaded', () => {
      const ball = document.getElementById('ball');
      let ballPosition = { x: 6, y: 1 }; // Starting grid position for the ball (column 6, row 1)

      // Hardcoded path: True means right, False means left
      const hardcodedPath = ["True", "False", "True", "False", "True"];

      // Function to move the ball based on the hardcoded path
      const moveBall = (path) => {
        let xPos = ballPosition.x;
        let yPos = ballPosition.y;
        let currentIndex = 0;

        const interval = setInterval(() => {
          if (currentIndex < path.length) {
            const moveRight = path[currentIndex] === "True";
            if (moveRight && xPos < 11) {
              xPos++; // Move right
            } else if (!moveRight && xPos > 1) {
              xPos--; // Move left
            }

            // Update ball position visually
            ball.style.left = `calc(${xPos * 8.5}% - 10px)`;
            ball.style.top = `calc(${yPos * 15}% - 10px)`;

            // Move down to the next row
            yPos++;
            currentIndex++;
          } else {
            clearInterval(interval); // Stop once the path is completed
          }
        }, 500); // Adjust the delay to control the speed of the ball movement
      };

      // Automatically start the ball movement with the hardcoded path
      moveBall(hardcodedPath);
    });
  </script>
